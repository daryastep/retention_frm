```python
# группирую доставленные заказы по ['month', 'weekday', 'customer_unique_id']

by_month = deliv_all.groupby(['month', 'weekday', 'customer_unique_id'], as_index=False) \
    .agg({'order_id': 'count'})
by_month
```

```python
# пишу фунцию, чтобы учесть количество недель в каждом месяце

def avg_by_week(x):
    if x['month'] in ['January', 'March', 'May', 'July', 'August', 'October', 'Decamber']:
        return x['order_id'] / 4.43
    elif x['month'] == 'February':
        return x['order_id'] / 4
    else:
        return x['order_id'] / 4.28
        
```

```python
# группирую по месяцам

bm = by_month.groupby(['month', 'customer_unique_id'], as_index=False).sum()\
    .sort_values('order_id', ascending=False)
bm
```

```python
# применяю фунцию
bm['avg_by_week'] = bm.apply(avg_by_week, axis=1)
```

## Сегментирование

    R=1  -  до 93 дня (примерно 3 месяцев) с момента последней покупки   
    F=1  -  0.1% клиентов с самым большим количеством покупок (более 3-х)
    M=1  -  1% клиентов с самыми большими чеками (более 1000)
    
    R=3  -  более 383 дней (примерно года) с момента последней покупки
    F=3  -  99% клиентов с самым маленьким количеством покупок (по одной)
    M=3  -  95% клиентов с самыми маленькими чеками (менее 419)
    

    
    R=2, F=2, M=2  - остальные клиенты со средним кол-вом покупок (от 2 до 3), 
    средними чеками (примерно 400 до 1000) и от 93 до 383 дней с момента последней покупки


```python
#  функция, чтобы сегментировать данные

def set_score(val, var, perc_1, perc_2):
    if val < perc_1:
        return 3 if var != 'R' else 1
    elif val < perc_2:
        return 2
    else:
        return 1 if var != 'R' else 3
    
# применяю функцию и создаю столбцы с R, F, M

recency_quantiles = for_rfm['days_since'].quantile(quantiles_r)
for_rfm['R'] = for_rfm['days_since'].apply(set_score, args=('R', recency_quantiles.iloc[0], recency_quantiles.iloc[1]))

frequency_quantiles = for_rfm['purchase_amount'].quantile(quantiles_f)
for_rfm['F'] = for_rfm['purchase_amount'].apply(set_score, args=('F', frequency_quantiles.iloc[0], frequency_quantiles.iloc[1]))

monetary_quantiles = for_rfm['purchase_sum'].quantile(quantiles_m)
for_rfm['M'] = for_rfm['purchase_sum'].apply(set_score, args=('M', monetary_quantiles.iloc[0], monetary_quantiles.iloc[1]))

for_rfm
```

```python
#  создаю колонку с RFM
for_rfm['RFM'] = for_rfm.apply(lambda row: f"{row['R']}{row['F']}{row['M']}", axis=1)
for_rfm
```
